```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import lightgbm as lgb
from sklearn import metrics, calibration

from sklearn.preprocessing import binarize, normalize
#from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
import itertools
from reporter import Reporter
from collections import OrderedDict
```

```python
data = pd.read_csv("creditcard.csv")
y_train = data.Class
X_train = data.drop(['Class','Time'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.4, random_state=42)
```

```python
X_train.shape , X_test.shape
```

```python
lgb_params = {
             'n_jobs': 5, 'two_round': True, 'is_unbalance': True,
             'boosting_type': 'gbdt', 'learning_rate': 0.01, 'min_child_samples': 10,
             'min_child_weight': 5, 'n_estimators':1000, 'num_leaves':21,
             'max_depth': 6
}

```

```python
clf = lgb.LGBMClassifier(**lgb_params, random_state = 10)
clf.fit(X_train, y_train,
        eval_set=[(X_test, y_test)],
        eval_metric='auc',
        early_stopping_rounds=100,
        verbose=False)
```

```python
rep = Reporter()
rep.put('Treino', y_train, clf.predict_proba(X_train)[:,1])
rep.put('Teste', y_test, clf.predict_proba(X_test)[:,1])
rep.report()
```
